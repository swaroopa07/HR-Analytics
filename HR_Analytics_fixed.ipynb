```python
df["SalaryBand"] = pd.qcut(
    df["MonthlyIncome"],
    q=4,
    labels=["Low", "Medium", "High", "Very High"]
)
```

```python
df["Experience"] = pd.cut(
    df["YearsAtCompany"],
    bins=[0, 5, 10, 20, 40],
    labels=["Associate", "Analyst", "Senior Analyst", "Manager"],
    include_lowest=True
)
```

```python
pd.crosstab(df["Experience"], df["SalaryBand"])
```

Employees who are unhappy with their work environment are more likely to leave than employees who are simply less satisfied with workplace relationships.

```python
df.groupby("NumCompaniesWorked")["PercentSalaryHike"].mean()
```

```python
df.groupby("NumCompaniesWorked")["MonthlyIncome"].mean()
```

“I analyzed the relationship between number of companies worked and compensation metrics. Salary hike percentages remained relatively constant across all groups, indicating that increments are likely policy-driven rather than based on job mobility. However, monthly income shows a mild increase for employees with moderate job-switching history (around 2–7 companies), suggesting that controlled mobility may contribute to higher earning potential, although the relationship is not strictly linear.”

```python
from sqlalchemy import create_engine

server = r'localhost\SQLEXPRESS'
database = 'HR_Analytics'   

connection_string = (
    f"mssql+pyodbc://@{server}/{database}"
    "?driver=ODBC+Driver+17+for+SQL+Server"
    "&trusted_connection=yes"
)

engine = create_engine(connection_string)

print("Connected successfully")
```

```python
# Upload dataframe as SQL table
df.to_sql(
    name='IBM_Data',
    con=engine,
    if_exists='replace',
    index=False
)

print("IBM data uploaded to SQL Server successfully.")
```
